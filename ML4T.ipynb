```python
import spacy
import pandas as pd
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
```

```python
sample_text = 'Apple is looking at buying U.K. startup for $1 billion'
doc = nlp(sample_text)
```

```python
pd.DataFrame([[t.text, t.lemma_, t.pos_, t.tag_, t.dep_, t.shape_, 
               t.is_alpha, t.is_stop]
              for t in doc],columns=['text', 'lemma', 'pos', 'tag', 'dep', 'shape', 'is_alpha', 'is_stop'])
```

```python
options = {'compact': True, 'bg': 'white',
           'color': 'black', 'font': 'Source Sans Pro', 'notebook': True}
displacy.render(doc, style='dep', options=options, jupyter=True)
```

```python
spacy.explain("VBZ")
```
